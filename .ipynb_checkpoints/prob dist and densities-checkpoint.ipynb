```python
import collections
```

```python
testlist = [1,2,3,4,6,9,9,9]
```

```python
c = collections.Counter(testlist)
```

```python
print(c)
```

```python
count_sum = sum(c.values())
```

```python
print(count_sum)
```

```python
for a,b in c.items():
    print('frequency of {} is {}'.format(str(a), str(float(b)/count_sum)))
```
