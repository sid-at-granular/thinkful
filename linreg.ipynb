```python
import pandas as pd
loansData = pd.read_csv('https://github.com/Thinkful-Ed/curric-data-001-data-sets/raw/master/loans/loansData.csv')
```

```python
loansData.head(50)
```

```python
loansData['Interest.Rate'] = loansData['Interest.Rate'].map(lambda x: x[:-1])
```

```python
loansData['Interest.Rate']
```

```python
loansData['Loan.Length'] = loansData['Loan.Length'].map(lambda x: x[:-7])
```

```python
loansData['Loan.Length'].head()
```

```python
loansData['FICO.Score'] = loansData['FICO.Range'].map(lambda x: x.split('-')[0])
```

```python
loansData.head()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

loansData['FICO.Score'] = loansData['FICO.Score'].map(lambda x: int(x))
```

```python
import numpy as np
import statsmodels.api as sm
```

```python
intrate = loansData['Interest.Rate'].map(lambda x: float(x))
loanamt = loansData['Amount.Requested']
fico = loansData['FICO.Score']
```

```python
intrate.head()
```

```python
y = np.matrix(intrate).transpose()
```

```python
x1 = np.matrix(fico).transpose()
x2 = np.matrix(loanamt).transpose()
```

```python
x = np.column_stack([x1,x2])
```

```python
X = sm.add_constant(x)
model = sm.OLS(y,X)
f = model.fit()
```

```python
f.summary()
```
